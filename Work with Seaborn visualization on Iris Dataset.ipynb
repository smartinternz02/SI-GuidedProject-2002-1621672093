```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
print(sns.get_dataset_names())
```

```python
dt = sns.load_dataset("iris")
```

```python
dt.head()
```

```python
plt.scatter(dt.sepal_length, dt.petal_length)
```

```python
sns.relplot(data=dt,x="sepal_length",y="petal_length",hue="species")
```

```python
sns.distplot(dt.sepal_length,kde=True)
```

```python
sns.distplot(dt.sepal_length,kde=False, bins=40)
```

```python
sns.boxplot(data=dt,x="sepal_length",y="petal_length",hue="species", width=2)
```

```python
sns.countplot(data=dt,x="species")
```

```python
sns.jointplot(data=dt,x="sepal_length",y="petal_length")
```

```python
sns.barplot(data=dt,x="sepal_length",y="petal_length",hue="species")
```

```python
sns.heatmap(dt.corr(),annot=True)
```

```python
sns.pairplot(data=dt,hue="species",kind="scatter",palette="husl")
```
